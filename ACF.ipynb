imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.stattools import acf
```

SPX data

```python
df = pd.read_csv("merged.csv")
data = df[df["SPX"] == 1]["Adj Close"]
```

plot ACF

```python
plot_acf(data, lags=20)
plt.title("SPX Autocorrelation")
plt.xlabel("Lags")
plt.ylabel("Auto-Corr")
plt.show() 
```

```python
acf_values = acf(data, nlags=20)
print(acf_values)
```

movie data

```python
data = df["Box Office"]

plot_acf(data, lags=20)
plt.title("Box Office Revenue Autocorrelation")
plt.xlabel("Lags")
plt.ylabel("Auto-Corr")
plt.show() 

acf_values = acf(data, nlags=20)
print(acf_values)
```
